```python
count = 0

for i in range(0,train.shape[0]):
#     cam = cv2.VideoCapture("C:\\Users\\Admin\\PycharmProjects\\project_1\\openCV.mp4")
    videoFile = train['video_name'][i]
    print(videoFile)
    cam = cv2.VideoCapture(videoFile.split(' ')[0])
 
    # frame 
    currentframe = 0
  
    while(True): 
        # reading from frame 
        ret,frame = cam.read() 
        if ret:
            
            if count%300 == 0 :
                
            # if video is still left continue creating images 

                filename = 'train1/'+videoFile.split('/')[2]+"_frame%d.jpg" % count;
                print ('Creating...' + filename) 
  
            # writing the extracted images 
                cv2.imwrite(filename, frame) 
  
            # increasing counter so that it will 
            # show how many frames are created 
                currentframe += 1
            count+=1
        else: 
            break
  
    # Release all space and windows once done 
    cam.release() 
    
```

```python
images = glob("train1/*.jpg")
train_image = []
train_class = []

for i in range(len(images)):
  
    train_image.append(images[i].split('\\')[1])
    
    train_class.append(images[i].split('\\')[1].split('_')[0])

train_data = pd.DataFrame()
train_data['image'] = train_image
train_data['class'] = train_class
train_data.to_csv('train_new.csv',header=True,index=False)
```
